# Linear Discriminant Analysis (LDA)
## PCA vs LDA
### PCA - No class separation is taken in consideration. It is collective and overall Principal components are considered 
### LDA - Class separation is maintained and is important. Totally different from PCA 


```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

## Acessing the Data

```python
wine = pd.read_csv('https://gist.githubusercontent.com/tijptjik/9408623/raw/b237fa5848349a14a14e5d4107dc7897c21951f5/wine.csv')
```

```python
wine
```

```python
X = wine.drop(['Wine'],axis=1)
```

```python
X
```

```python
X.shape
```

```python
y = wine['Wine']
```

```python
y
```

```python
y.unique()
```

## Standardisation of Features 

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_scaled = sc.fit_transform(X)
X_scaled
```

## Building LDA Model

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
# n_components = min(n_classes, n_features)
# Here n_classes = 3 ( 3 types of wine - 1,2,3)
# And n_features = 13 (13 columns)
# so n_components = min(3-1,13) = min(2,13) = 2
# So we take a HYPERPLANE of Dimension 2 
```

```python
lda = LinearDiscriminantAnalysis(n_components=2)
```

```python
type(lda)
```

## Transforming the Data

```python
X_lda = lda.fit_transform(X_scaled, y) # y done as y is important for in class, between class and all
X_lda
```

```python
X_lda.shape
```

## Visualisation of the Data 

```python
plt.scatter(X_lda[:,0], X_lda[:,1], c=y)
```

# Locally Linear Embedding (LLE)

### LLE is used whenever projection fails. When data points cannot be projected onto a 2D frame. Eg: Swiss Roll example

### 1. Manifold Learning 
### 2. Focuses on nearby points of each data point
### 3. Distance is locally preserved 

### ====> We want to preserve the local nature of the data points in LLE 

### For each data point, identify k neighbours. Represent this point as linear combination of its neighbours using inital random weights. We get a weight matrix W 

```python
from sklearn.datasets import make_swiss_roll
```

```python
X,y = make_swiss_roll(n_samples=1000, random_state=100)
X

# Swiss Roll is 3D so machine also shows 3D 
```

```python
y
```

## Building the Model

```python
from sklearn.manifold import LocallyLinearEmbedding
```

```python
lle = LocallyLinearEmbedding(n_neighbors=5, random_state=100)
```

```python
type(lle)
```

## Transforming the Data 

```python
X_lle = lle.fit_transform(X)
```

```python
X_lle
```

## Visualisation 

```python
plt.scatter(X_lle[:,0], X_lle[:,1], c=y)
```

## Changing the n_neighbors to 10 

```python
lle = LocallyLinearEmbedding(n_neighbors=10, random_state=100)
X_lle = lle.fit_transform(X)
plt.scatter(X_lle[:,0], X_lle[:,1], c=y)
```

## Changing the n_neighbors to 15

```python
lle = LocallyLinearEmbedding(n_neighbors=15, random_state=100)
X_lle = lle.fit_transform(X)
plt.scatter(X_lle[:,0], X_lle[:,1], c=y)
```
