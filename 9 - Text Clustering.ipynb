# Text Clustering 

Text clustering, also known as document clustering or text segmentation, is a natural language processing technique that involves grouping similar documents or pieces of text together into clusters or categories. The main goal of text clustering is to identify groups of documents that share similar topics or themes based on the content of the text.

Text clustering can be used in a variety of applications, such as information retrieval, content analysis, document organization, and text classification. It can also help in identifying patterns and relationships between documents, which can be useful for tasks such as recommendation systems, sentiment analysis, and topic modeling.

The process of text clustering involves several steps, including pre-processing the text data by removing stop words, stemming, and tokenizing the text, defining similarity metrics to compare the text documents, selecting an appropriate clustering algorithm such as k-means, hierarchical clustering, or DBSCAN, and evaluating the results to ensure that the clusters are meaningful and useful

## Converting words to TFIDF vectors

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf = TfidfVectorizer()
```

```python
features = tf.fit_transform(data)
```

```python
features
```

```python
print(features)
```

```python
# The first number is the document number 
# The second number is the ID number for the particular word 
# The third number is the tfidf vectorizer 
```

## Clustering using K-means

```python
from sklearn.cluster import KMeans
SSD=[]
for k in range(1,10):
    kmeans=KMeans(n_clusters=k, random_state=10)
    kmeans.fit(features)
    SSD.append(kmeans.inertia_)  
plt.plot(range(1,10),SSD);
```

```python
from sklearn.metrics import silhouette_score
SS = []
for k in range(2,11): 
    kmeans = KMeans(n_clusters = k, random_state=10)
    kmeans.fit(features)
    SS.append(silhouette_score(features,kmeans.predict(features)))
plt.plot(range(2,11),SS);
```

Suppose no.of Clusters = 4 

## Building a Clustering Model

```python
kmeans = KMeans(n_clusters = 4, random_state = 10)
kmeans.fit(features)
clust_pred = kmeans.predict(features)
clust_pred
```

## Appending the DataFrame

```python
bbc['Cluster'] = kmeans.labels_
```

```python
bbc
```

## Dimensionality Reduction & Visualization

```python
from sklearn.manifold import TSNE 

tsne = TSNE(n_components=2, perplexity=30)
```

```python
features_tsne = tsne.fit_transform(features)
```

```python
print(features_tsne)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(features_tsne[:,0], features_tsne[:,1], c=bbc['Cluster'])
```
