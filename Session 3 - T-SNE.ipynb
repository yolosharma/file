## Create a Random Sample of 1000 from 70000

```python
np.random.seed(100)
sample = np.random.choice(X.shape[0],1000)      # select 1000 points randomly from 70000 where X.shape[0] = 70000
print(sample)
```

## Creating a Subset of 1000 samples of X & y 

```python
X1 = X.iloc[sample,:]  # Creating X1 which has all the columns(,:) of X for the indexes in samples 
X1.shape
```

```python
X1.head()
```

```python
y1 = y[sample]
y1.shape
```

```python
y1.head()
```

## Building the TSNE Model 

```python
from sklearn.manifold import TSNE 
```

```python
tsne = TSNE(n_components=2, perplexity=30)
```

```python
X_tsne = tsne.fit_transform(X1)
X_tsne.shape
```

```python
# The number of features have been reduced from 784 to 2 
```

```python
X_tsne[0]
```

## Plotting the transformed Data Points

```python
plt.scatter(X_tsne[:,0], X_tsne[:,1]);
```

```python
plt.scatter(X_tsne[:,0], X_tsne[:,1], c=y1);
```

```python
plt.scatter(X_tsne[:,0], X_tsne[:,1], c=y1.astype(float));
```

## Creating a DataFrame

```python
X_df = pd.DataFrame({'X0':X_tsne[:,0],
                    'X1':X_tsne[:,1],
                    'Label':y1})

X_df
```

```python
plt.figure(figsize=(15,12))
sns.lmplot(data=X_df, x='X0', y='X1',
          hue='Label');
```

```python
plt.figure(figsize=(15,12))
sns.lmplot(data=X_df, x='X0', y='X1',
          hue='Label', fit_reg=False);
```
