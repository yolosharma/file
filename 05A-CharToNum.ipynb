####Charachter (Categoric) To Numeric Conversion

```
# imports
# hides all warnings
import warnings
warnings.filterwarnings('ignore')
# import pandas for spark
import pyspark.pandas as ps
# pandas 
import pandas as pd
# numpy
import numpy as np
```

```
# convert string (categoric) dataframe column to numeric
"""
desc:
    CharTo  - removes Outliers from all cols in df except exclCols 
usage: 
    CHarToNum(df, colName, keyMap) 
params:
    df datarame, 
    colName should be charachter categoic
    keyMap dict containng map of char to num conversion  
returns:
    Nil
"""
def CharToNum(df, colName, keyMaps):
    # repl null with mode
    replVal = df[colName].mode()
    df[colName] = df[colName].fillna(replVal[0])
    # convet using map 
    df[colName] = df[colName].map(keyMaps)
    # convert to int
    df[colName] = df[colName].astype(int)
    return
```

```
# read csv

# just change location & name as applicable
# File location and type
file_location = "/FileStore/tables/test/insurance.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

```
# convert dataframe to pandas spark dataframe
psdf = ps.DataFrame(df)
print(psdf.head(5))
```

```
# info
print("\n*** Structure ***")
print(psdf.info())
```

```
# convert string / categoric to numeric
colName = 'gender'
print(colName.title())
print(psdf[colName].unique())
CharToNum(psdf, colName, {"female": 0, "male": 1})
print(psdf[colName].unique())
```

```
# convert string / categoric to numeric
colName = 'smoker'
print(colName.title())
print(psdf[colName].unique())
CharToNum(psdf, colName, {"no": 0, "yes": 1} )
print(psdf[colName].unique())
```

```
# convert string / categoric to numeric
colName = 'region'
print(colName.title())
print(psdf[colName].unique())
CharToNum(psdf, colName, {"northeast": 0, "northwest": 1,"southeast": 2, "southwest": 3} )
print(psdf[colName].unique())
```
